```python
Xroll = X1[0]
mapper = {}
for k in range(len(X2)):
    Yroll = np.array(X2[k])
    equality = equalityCheck(Xroll, Yroll)
    if(equality != 1):
        mapper[k] = equality
            
sortedMap = {k: v for k, v in sorted(mapper.items(), key=lambda item: item[1])}
closestRolls = []        

for key in sortedMap:
    if(len(closestRolls) == 10):
        break
    closestRolls.append(X2[key])
```

```python
print(np.array(closestRolls).shape)
```

```python
for i in range(len(closestRolls)):
    roll = reshape(closestRolls[i])
    roll = roll * 100
    pm = piano_roll_to_pretty_midi(roll, 100, 0)
    pm.write(pathToSave + "\\closest" + str(i) + ".mid")
```
