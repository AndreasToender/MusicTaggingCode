```python
from pydub import AudioSegment
import joblib
import os.path
import glob
```

```python
workdir = "C:\\Users\\toend\\Documents\\ITU\\Thesis"
path = os.path.join(workdir, "MP3files\\2018MP3Files")
pathToSave = os.path.join(workdir, "MP3files\\18Cuts")
sequenceLength = 30 #seconds
```

```python
def cut_mp3_files(mp3, saveTo):
    song = AudioSegment.from_mp3(mp3)
    songLength = int(len(song)/1000)
    for i in range(0, songLength - sequenceLength, sequenceLength):
        cut = song[i*1000:i*1000 + sequenceLength*1000]
        #MIGHT HAVE TO PAD
        cut.export(saveTo + '-' + str(i) + '.mp3', format="mp3")
```

```python
#joblib.Parallel(n_jobs=10, verbose=1)(joblib.delayed(cut_mp3_files)(mp3)for mp3 in glob.glob(os.path.join(path)))
```

```python
with os.scandir(path) as entries:
    for entry in entries:
        if entry.is_file():
            name = entry.name.replace(".mp3", "")
            filename = "18" + name
            saveTo = os.path.join(pathToSave, filename)
            print(saveTo)
            cut_mp3_files(entry, saveTo)
            
```
